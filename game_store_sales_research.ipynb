```python
def print_pearson(name='PS4'):
    print('–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞',name,'\n',
          '–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–π —Ä–µ–π—Ç–∏–Ω–≥–∞ –∫—Ä–∏—Ç–∏–∫–æ–≤ –∏ —Å—É–º–º–∞—Ä–Ω—ã—Ö –ø—Ä–æ–¥–∞–∂ ','\n',
            np.corrcoef(df[(df['platform']==name)&(df['critic_score']>0)]['critic_score'],
                    df[(df['platform']==name)&(df['critic_score']>0)]['world_sales']),'\n','\n',
          '–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–≥–æ –∏ —Å—É–º–º–∞—Ä–Ω—ã—Ö –ø—Ä–æ–¥–∞–∂ ','\n',
            np.corrcoef(df[(df['platform']==name)&(df['user_score']>0)]['user_score'],
                    df[(df['platform']==name)&(df['user_score']>0)]['world_sales']),'\n','\n',
           )
```

```python
def rating_influence(name = 'PS4'):
    print('–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞',name +': –ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–∏—Ä–æ–≤—ã—Ö –ø—Ä–æ–¥–∞–∂ –æ—Ç –æ—Ü–µ–Ω–∫–∏ –∫—Ä–∏—Ç–∏–∫–æ–≤.' )
    (df[(df['platform']==name)&(df['critic_score']>0)].pivot_table(index='critic_score',values='world_sales',aggfunc='mean')
       .plot(kind='barh', figsize=(10, 10), grid=True));
    plt.xlabel('–ü—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∫–æ–ø–∏–π, –º–ª–Ω')
    plt.show();
    data = df[(df['platform']==name)&(df['critic_score']>0)]
    sns.scatterplot( x="world_sales", y="critic_score",data=data)
    plt.show();

    print('–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞',name +': –ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–∏—Ä–æ–≤—ã—Ö –ø—Ä–æ–¥–∞–∂ –æ—Ç –æ—Ü–µ–Ω–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.' )
    (df[(df['platform']==name)&(df['user_score']>0)].pivot_table(index='user_score',values='world_sales',aggfunc='mean')
       .plot(kind='barh', figsize=(10, 10), grid=True));
    plt.show();
    plt.xlabel('–ü—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∫–æ–ø–∏–π, –º–ª–Ω')
    data = df[(df['platform']==name)&(df['user_score']>0)]
    sns.scatterplot( x="world_sales", y="user_score",data=data)
    plt.show();
```

```python
def top_sales_by_area(field_name='platform',y_name='na_sales', y_max=100, agg='sum'):
    if agg == 'sum':
        data=df.groupby(field_name)[y_name].sum().sort_values(ascending=False).reset_index().head()
    elif agg == 'mean':
        data=df.groupby(field_name)[y_name].mean().sort_values(ascending=False).reset_index().head()
    sns.set(rc={'figure.figsize':(5,5)})
    sns.barplot(
        x=field_name,
        y=y_name,
#        hue='platform',
        data=data
    ).set(ylim=(0,y_max))
    plt.show();
```

```python
limit =  [100, 140, 50]
for area in ['na_sales', 'eu_sales', 'jp_sales']:
    for lim in limit:
        top_sales_by_area(field_name='platform', y_name=area, y_max=lim, agg='sum')
        limit.pop(0)
        break;
```

```python
for area in ['na_sales', 'eu_sales', 'jp_sales']:
    top_sales_by_area(field_name='platform', y_name=area, y_max=.4, agg='mean')
```

–ü—Ä–æ–≤–µ—Ä–∏–º –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ "—Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ". –î–ª—è —ç—Ç–æ–≥–æ —Å—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –≥—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: "—Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC —Ä–∞–≤–Ω—ã" –∏ –ø–æ–ø—Ä–æ–±—É–µ–º –µ–µ –æ–ø—Ä–æ–≤–µ—Ä–≥–Ω—É—Ç—å. –ü—Ä–µ–∂–¥–µ —á–µ–º –ø—Ä–æ–≤–æ–¥–∏—Ç—å –ø—Ä–æ–≤–µ—Ä–∫—É –≥–∏–ø–æ—Ç–µ–∑—ã –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –¥–≤—É—Ö –≥–µ–Ω–µ—Ä–∞–ª—å–Ω—ã—Ö —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–µ–π –ø–æ –≤–∑—è—Ç—ã–º –∏–∑ –Ω–∏—Ö –≤—ã–±–æ—Ä–∫–∞–º, –ø–æ—Å—á–∏—Ç–∞–µ–º —Ä–∞–∑–º–µ—Ä—ã –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –≤—ã–±–æ—Ä–æ–∫
