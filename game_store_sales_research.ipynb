 # Исследование продаж компьютерных игр интернет-магазина "Стримчик"

Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xboxили PlayStation). Предоставлены данные до 2016 года. Необходимо выявить определяющие успешность игры закономерности и выявить потенциально популярный продукт для планирования рекламных кампаний на 2017-й год. В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

  - в полях с названием игры, жанра и возрастным рейтингом - значением "unknown"

```python
df['genre'] = df['genre'].fillna('unknown')
df['rating'] = df['rating'].fillna('unknown')
df['name'] = df['name'].fillna('unknown')

df['year_of_release'] = df['year_of_release'].fillna(-1)
df['critic_score'] = df['critic_score'].fillna(-1)

df['user_score'] = df['user_score'].replace('tbd',-1)
df['user_score'] = df['user_score'].fillna(-1)
```

Преобразуем типы в соответствии с принятыми в п 1.1 решениями

```python
df['user_score'] = df['user_score'].astype('float')
df['critic_score'] = df['critic_score'].astype('int')
df['year_of_release']  = df['year_of_release'].astype('int')
df.head()
```

Рассчитаем суммарные продажи по всем регионам и запишем их в столбец _world_sales_

```python
df['world_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
df.head()
```

Рассмотрим для игр, разработанных на этих платформах, как изменялись от года к году суммарные продажи

```python
platforms = (df[(df['year_of_release']>0) & df['platform'].isin(pl_names)]
             .groupby(['platform', 'year_of_release'])['na_sales','eu_sales','jp_sales','other_sales','world_sales']
             .sum()
             .reset_index()
            )

platforms.head(15)
```

```python
def plotting_sales(data=platforms, name='DS', y='world_sales'):
    print('Мировые продажи по годам для игр, выпущенных на платформе', name)
    fig = sns.barplot()
    plt.grid(visible=True, which='major')
    sns.set(rc={'figure.figsize':(6,6)})
    fig = sns.barplot(
                      data=data[data['platform']==name],
                      x="year_of_release",
                      y='world_sales')
    plt.show()

for pl_name in pl_names:
    plotting_sales(name = pl_name)
```

Из гистограмм выше можно сделать вывод, что срок жизни платформы - в среднем 10 лет. Платформы, являющиеся "наследниками" известных популярных "родителей", такие как PS2 и PS3, а также Wii и DS производства nintendo, ожидаемо стартуют с более высоких показателей в первые годы после выпуска в сравнении с менее известными на момент выпуска конкурентами.

Однако, вне зависимости от этих показателей, среднее время жизненного цикла платформ примерно равно - 10 лет, причем в последние пару "лет жизни" количество продаж минимально. В связи с этим для построения прогноза на 2017 год необходимо рассматривать только игры года выпуска 2009 и позже.

## Шаг 5. Проверка гипотез

В качестве исходных данных используются выборки из двух больших массивов данных: геймеры платформ Xbox One и PC в первом случае, и геймеры жанров Action и Sports  - во втором. Так как необходимо проверить гипотезы касаемо о средних значений месячного платежа для абонентов, будем применять метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам.

### "Cредние пользовательские рейтинги платформ Xbox One и PC одинаковые"

Проверим гипотезу о том, что "средние пользовательские рейтинги платформ Xbox One и PC одинаковые". Для этого сформулируем гулевую гипотезу: "средние пользовательские рейтинги платформ Xbox One и PC равны" и попробуем ее опровергнуть. Прежде чем проводить проверку гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам, посчитаем размеры предоставленных выборок

```python
print(
      df[(df['user_score']>0)&(df['platform']=='XOne')]['user_score'].count(),
      df[(df['user_score']>0)&(df['platform']=='PC')]['user_score'].count()
     )
```

```python
alpha= .05
results = st.ttest_ind(df[(df['user_score']>0)&(df['platform']=='XOne')]['user_score'],
                       df[(df['user_score']>0)&(df['platform']=='PC')]['user_score'],
                       equal_var=False # так как выборки разных размеров
                      )

print(' p-значение:', results.pvalue,'\n',
      'Средний пользовательский рейтинг платформы Xbox One:',
      round(df[(df['user_score']>0)&(df['platform']=='XOne')]['user_score'].mean(), 1),'\n',
      'Средний пользовательский рейтинг платформы PC:',
      round(df[(df['user_score']>0)&(df['platform']=='PC')]['user_score'].mean(), 1)
     )

if results.pvalue<alpha:
    print(' Отвергаем нулевую гипотезу')
else:
    print(' Не получилось отвергнуть нулевую гипотезу')
```

#### Вывод
Гипотеза не подтвердилась: средние пользовательские рейтинги платформ Xbox One и PC различаются

###  "Cредние пользовательские рейтинги жанров Action и Sports разные"

Проверим гипотезу о том, что "средние пользовательские рейтинги жанров Action и Sports разные". Для этого сформулируем нулевую гипотезу: "средние пользовательские рейтинги жанров Action и Sports равны" и попробуем ее опровергнуть. И снова предвапительно посчитаем размеры предоставленных выборок

```python
print(
      df[(df['user_score']>0)&(df['genre']=='Action')]['user_score'].count(),
      df[(df['user_score']>0)&(df['genre']=='Sports')]['user_score'].count()
     )
```

```python
alpha= .05
results = st.ttest_ind(df[(df['user_score']>0)&(df['genre']=='Action')]['user_score'],
                       df[(df['user_score']>0)&(df['genre']=='Sports')]['user_score'],
                       equal_var=False
                      )

print(' p-значение:', results.pvalue,'\n',
      'Средний пользовательский рейтинг игр жанра Action:',
      round(df[(df['user_score']>0)&(df['genre']=='Action')]['user_score'].mean(), 1),'\n',
      'Средний пользовательский рейтинг игр жанра Sports:',
      round(df[(df['user_score']>0)&(df['genre']=='Sports')]['user_score'].mean(), 1)
     )

if results.pvalue<alpha:
    print(' Отвергаем нулевую гипотезу')
else:
    print(' Не получилось отвергнуть нулевую гипотезу')
```

#### Вывод
Нулевая гипотеза опровергнута, а это значит, что исходная гипотеза верна: средние пользовательские рейтинги жанров Action и Sports разные
